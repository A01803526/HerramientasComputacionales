```python
# Medidas de tendencia central
print(f"Promedio: {df['Confirmed'].mean()}")
print(f"Mediana: {df['Confirmed'].median()}")
print(f"Moda: {df['Confirmed'].mode()}")
# Medidas de dispersión
print(f"Desviación estándar: {df['Confirmed'].std()}")
print(f"Varianza: {df['Confirmed'].var()}")
# Medidas de posicionamiento
print(f"Mínimo: {df['Confirmed'].min()}")
print(f"Q1: {df['Confirmed'].quantile(0.25)}")
print(f"Q2: {df['Confirmed'].quantile(0.50)}")
print(f"Q3: {df['Confirmed'].quantile(0.75)}")
print(f"Máximo: {df['Confirmed'].max()}")
```

```python
columnas_numéricas = []
columnas_no_numéricas = []
for col in df.columns:
    if df[col].dtype == "object":
        columnas_no_numéricas.append(col)
    else:
        columnas_numéricas.append(col)
print(f"Columnas numéricas: {columnas_numéricas}")
print(f"Columnas no numéricas: {columnas_no_numéricas}")
```

```python
for col in columnas_numéricas:
    df[col].plot.box()
    plt.show()
```

```python
for col in columnas_numéricas:
    datos = df[col].replace([np.inf, -np.inf], np.nan).dropna()
    if len(datos) > 0:
        datos.plot.kde(title=col)
        plt.show()
```
